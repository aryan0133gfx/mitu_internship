```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.array([1,2,3,4,5,6])
y= np.array([2,3,4,6,7,8])
z= np.array([8,4,5,8,5,7])
```

### Line plot

```python
plt.title('Line plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(2,10,2))
plt.yticks(range(2,10,2))
plt.plot(x,y,marker = '*',mfc = 'red',color = 'black',ls = ':',ms=7);
plt.grid()
```

```python
plt.plot(x,z);
```

```python
plt.title('Line plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(2,10,2))
plt.yticks(range(2,10,2))
plt.plot(x,y,marker = 'o',mfc = 'yellow',color = 'black',ls = ':',ms=7,label = '2023');
plt.plot(x,z,marker = '*',mfc = 'red',color = 'black',ls = ':',ms=7,label = '2024');
plt.grid()
plt.legend();
```

```python
# subplot
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.title('Line plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(2,10,2))
plt.yticks(range(2,10,2))
plt.plot(x,y,marker = '*',mfc = 'red',color = 'black',ls = ':',ms=7);
plt.grid()

plt.subplot(2,2,2)
plt.title('Line plot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.xticks(range(2,10,2))
plt.yticks(range(2,10,2))
plt.plot(x,y,marker = '*',mfc = 'red',color = 'black',ls = ':',ms=7);
plt.grid()
```

### Bar plot

### box plot

```python
marks = [67,65,62,64,55,53,75,91,23]
```

```python
plt.boxplot(marks);
```

```python
q3  = np.percentile(marks,75)
q2 = np.percentile(marks,25)
q3,q2
```

```python
iqr = q3 -q2
iqr
```

```python
ub = q3 +(1.5 * iqr)
lb = q2-(1.5 * iqr)
```
